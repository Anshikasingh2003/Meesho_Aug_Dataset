```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
data=pd.read_csv(r"C:\Users\anshi\OneDrive\Desktop\meesho\meesho Orders Aug.csv")
data.head()
```

```python
data.drop(["Supplier Discounted Price (Incl GST and Commision)"],axis=1,inplace=True)
```

```python
data.columns=["Status","Order No." ,"Date" , "State","Product Name","SKU","Size","Quantity","Price"]
```

```python
plt.pie(data["Status"].value_counts(),autopct="%.0f%%",labels=data["Status"].unique(),radius=0.8)
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.countplot(x=data["Size"])
plt.show()
```

```python
data["Size"].value_counts()
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x=data["Size"] , y=data["Price"],palette="colorblind")
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x=data["Size"].unique() , y= data.groupby(["Size"])["Price"].sum())
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x=data["Status"].unique() , y= data.groupby(["Status"])["Price"].sum())
```
